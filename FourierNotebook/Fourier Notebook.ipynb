```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
data = pd.read_csv("data.csv", delimiter  = ";")
```

```python
MagX = data["MagX"]
MagY = data["MagY"]
MagZ = data["MagZ"]
roll = data["roll"]
pitch= data["pitch"]
yaw = data["yaw"]
AccX = data["AccX"]
AccY = data["AccY"]
AccZ = data["AccZ"]
GyrX = data["GyrX"]
GyrY = data["GyrY"]
GyrZ = data["GyrZ"]
```

```python
import numpy as np
from numpy.fft import fft, ifft
import matplotlib.pyplot as plt
# Plot the fft of the time vs. temperature graph:
sr = 10000 # Sampling rate
X = fft(roll)
N = len(X)
n = np.arange(N)
T = N/sr
freq = n/T 

fig, ax = plt.subplots(3, 2)
fig.dpi = 400
ax[0,0].plot(roll)
ax[0,1].plot(freq, np.abs(X))

X = fft(pitch)
N = len(X)
n = np.arange(N)
T = N/sr
freq = n/T 
ax[1,0].plot(pitch)
ax[1,1].plot(freq, np.abs(X))

X = fft(yaw)
N = len(X)
n = np.arange(N)
T = N/sr
freq = n/T 
ax[2,0].plot(yaw)
ax[2,1].plot(freq, np.abs(X))
ax[0, 0].set_title("Magnitude of Roll vs Time(s)")
ax[0, 1].set_title("FFT Amplitude |Roll| vs Frequency(Hz)")
ax[1, 0].set_title("Magnitude of Pitch vs Time(s)")
ax[1, 1].set_title("FFT Amplitude |Pitch| vs Frequency(Hz)")
ax[2, 0].set_title("Magnitude of Yaw vs Time(s)")
ax[2, 1].set_title("FFT Amplitude |Yaw| vs Frequency(Hz)")
plt.subplots_adjust(left=0.1, 
                    bottom=0.1,  
                    right=0.9,  
                    top=0.9,  
                    wspace=1,  
                    hspace=1) 
 
plt.show()
```
